```python
x = np.linspace(0,10,20)
y = x*x
z = x+y
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo Graph")
```

```python
plt.plot(x,x)
```

```python
plt.plot(x,z)
```
