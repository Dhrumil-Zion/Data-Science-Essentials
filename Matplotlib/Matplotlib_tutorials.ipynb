```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt



## https://matplotlib.org/
```

```python
x = np.linspace(0,10,20)
y = x*x
z = x+y
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo Graph")
```

```python
plt.plot(x,x)
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y,x,x,z,x)
```

```python
plt.subplot(3,2,1)
plt.plot(x,y,x,x,z,x)
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,3)
plt.plot(x,x)
plt.subplot(3,2,4)
plt.plot(z,x)
plt.subplot(3,2,5)
plt.plot(x,y,x,x,z,x)
plt.subplot(3,2,6)
plt.plot(x,y,x,x,z,x)
plt.tight_layout(pad=2.0, h_pad=None, w_pad=None)
```

```python
plt.fill(x,y,label = "Legend")
plt.legend(loc =2) ## loc = 2,3,4 ... 9,10
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y,marker='o',markersize = 15,markeredgecolor = 'black',markeredgewidth = 2)
```

```python
figgg = plt.figure(figsize=(10,10))
dd = figgg.add_axes([0.1,0.1,1,1])
dd.plot(x,y,color='g',linewidth=5,alpha=0.5,linestyle='--')     # alpha is for opacity of line 
ff = figgg.add_axes([0.2,0.2,0.5,0.5])
ff.plot(x,x,color='r',linewidth=2,marker='o',markersize=7,markerfacecolor='b')
```
